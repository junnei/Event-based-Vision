# **이벤트 카메라**
**Event-based Vision**

- 빛의 밝기 변화를 측정하여 정보를 저장하는 새로운 개념의 생체 모방형 카메라.

기존 카메라에 비해 다음과 같은 장점을 가지고 있음.

- 높은 동적 범위(High Dynamic Range)
- 낮은 전력 소모량(Low Power)
- 높은 시간 분해능력(High Temporal Resolution)

<br>

위의 장점에 따른 대표적인 Positive Effects 는 다음과 같음.

- 조도에 따른 변화가 비교적 적음(No Saturation)
- 동작흐림, 즉 모션블러 현상이 적음(No motion Blur)

zip 파일을 풀면 다음 파일들이 존재한다.

- events.txt
 - 한 줄에 하나의 이벤트값 [시간 x y 극성(polarity)]<br>

- images.txt
 - 한 줄에 하나의 이미지 경로 (시간 파일이름)
 - One image reference per line (timestamp filename)

- images/00000001.png
 - images.txt 에 쓰여있는 이미지
 - Images referenced from images.txt

 - One measurement per line (timestamp ax ay az gx gy gz)
 
- groundtruth.txt
 - 각 줄마다 Ground truth 값 (시간 px py pz qx qy qz qw)
 - One ground truth measurements per line (timestamp px py pz qx qy qz qw)

- calib.txt
 - 카메라 파라미터 (fx fy cx cy k1 k2 p1 p2 k3)
 - Camera parameters (fx fy cx cy k1 k2 p1 p2 k3)

# **이벤트 데이터 시각화**
**Visualizing Event-based Data**

````
# 쉬운 이해를 위해 직관적인 함수를 사용하였기 때문에 코드가 비효율적일 수 있음.
import numpy as np

#테스트할 이미지가 찍힌 시간(타임스태프) 가져오기
f = open("data/images.txt", 'r')
line = f.readline()
time = line.split(" ")[0]
print(time)
f.close()

#이벤트 데이터 저장공간
img_pos = np.zeros((240 , 180), dtype="float32")
img_neg = np.zeros((240 , 180), dtype="float32")


#이미지가 찍힌 시간(타임스태프)에 상응하는 이벤트 데이터 가져오기
f = open("data/events.txt", 'r')
lines = f.readlines()

threshold=False
while (threshold==False):
    for line in lines:
        line_data = line.split(" ")
        if (line_data[0] > time):
            threshold = True
            break
        if (line_data[3][:-1]=="1"):
            np.add.at(img_pos, (int(line_data[1]),int(line_data[2])),1)
        #else if (line_data[3][:-1]=="0")
        else:
            np.add.at(img_neg, (int(line_data[1]),int(line_data[2])),1)
f.close()

histogram = img_pos - img_neg

print(histogram)
````

````
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12, 9))

#Positive One
pos = fig.add_subplot(221)

find = np.where(img_pos!=0)
for i in range(np.shape(find)[1]):
    pos.scatter(find[0][i],find[1][i],s=abs(img_pos[find[0][i], find[1][i]]), color='green')

pos.set_xlim([0, 240])
pos.set_ylim([180, 0])
pos.set_title("Positive Representation")

#Negative One
neg = fig.add_subplot(222)

find = np.where(img_neg!=0)
for i in range(np.shape(find)[1]):
    neg.scatter(find[0][i],find[1][i],s=abs(img_neg[find[0][i], find[1][i]]), color='red')

neg.set_xlim([0, 240])
neg.set_ylim([180, 0])
neg.set_title("Negative Representation")

#Merge One
mer = fig.add_subplot(223)

mer.set_xlim([0, 240])
mer.set_ylim([180, 0])
mer.set_title("Event Representation")

#Histogram
ax = fig.add_subplot(224)

ax.set_facecolor('grey')

find = np.where(histogram!=0)
for i in range(np.shape(find)[1]):
    ax.scatter(find[0][i],find[1][i],s=abs(histogram[find[0][i], find[1][i]]), color=('black' if histogram[find[0][i], find[1][i]]<0 else 'white'))
    mer.scatter(find[0][i],find[1][i],s=abs(histogram[find[0][i], find[1][i]]), color=('red' if histogram[find[0][i], find[1][i]]<0 else 'green'))

ax.set_xlim([0, 240])
ax.set_ylim([180, 0])
ax.set_title("Greyscale Representation")

plt.show()

print("Original Image")
display.Image("data/"+image)
````

3D 시각화

````
all_ts=[]
all_x=[]
all_y=[]
all_p=[]

f = open("data/events.txt", 'r')

num_events=50000
lines = f.readlines()
for line in lines[:num_events]:
    line_data = line.split(" ")
    all_ts.append(float(line_data[0]))
    all_x.append(int(line_data[1]))
    all_y.append(int(line_data[2]))
    all_p.append(int(line_data[3][:-1]))
f.close()
````

````
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np 
import matplotlib

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111, projection='3d') # Axe3D object

frame_start = 0
sample_size = 50000
for i in range(frame_start,frame_start+sample_size):
    if (all_p[i]==1):
        ax.scatter(all_ts[i],all_x[i],all_y[i],s= 1, color='green')
    else:
        ax.scatter(all_ts[i],all_x[i],all_y[i],s= 1, color='red')

ax.set_ylim([0, 240])
ax.set_zlim([180, 0])

ax.set_title("Event data : 3D View")

plt.show()
````
